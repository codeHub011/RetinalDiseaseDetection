```python
INPUT_SHAPE = (224,224,3)
```

```python
mobilenet = tf.keras.applications.MobileNetV3Large(
    input_shape=INPUT_SHAPE,
    alpha=1.0,
    minimalistic=False,
    include_top=True,
    weights="imagenet",
    input_tensor=None,
    classes=1000,
    pooling=None,
    dropout_rate=0.2,
    classifier_activation="softmax",
    include_preprocessing=True,
   
)
```

```python
model = tf.keras.models.Sequential()
```

```python
model.add(tf.keras.Input(shape=INPUT_SHAPE))
```

```python
model.add(mobilenet)
```

```python
model.add(tf.keras.layers.Dense(units = 4,activation="softmax"))
```

```python
metrics_list = ["accuracy"]
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),loss="categorical_crossentropy",metrics=metrics_list)
```

```python
from sklearn.metrics import f1_score

y_true = []
y_pred = []

for images, labels in test_set:
    preds = model.predict(images)
    y_pred.extend(preds.argmax(axis=1))
    y_true.extend(labels.numpy().argmax(axis=1))

f1 = f1_score(y_true, y_pred, average='weighted')
print("F1 Score:", f1)
```
